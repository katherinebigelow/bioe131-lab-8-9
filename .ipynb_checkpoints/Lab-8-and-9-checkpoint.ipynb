## Running SPAdes

_Why do we expect short reads to produce a more fragmented assembly than long reads? Why does a single-molecule sequencing like PacBio have a higher error rate than Illumina?_

We expect short reads to produce a more fragmented assembly than long reads because since the reads are shorter, that means that they are smaller in size. The long reads are bigger in size and so they are more contiguous and result in a less fragmented assembly.  


A single-molecule sequencing like PacBio has a higher error rate than Illumina because PacBio uses longer reads compared to Illumina so this means that if I want to run through the same data more than once, this becomes inefficient. Illumina uses shorter reads compared to PacBio so this means that Illumina can create multiple outputs for one read since it doesn't take as long as long reads.

The command I use to run is: spades.py -1 /data/lab8/illumina_reads_R1.fastq -2 /data/lab8/illumina_reads_R2.fastq --pacbio /data/lab8/pacbio_reads.fastq  -o /home/be131-13/lab8/spaoutput -t 1

_Why is N50 is useful statistic to calculate? Why not just list the mean or median contig length? In your assembly, are scaffolds longer than contigs, or are scaffolds approximately equal to contigs? _

N50 is useful because it is a weighted median statistic. It helps tell you that around 50% of the assembly is contained in the contigs or scaffolds file that is large than or equal to this value. In my assembly, the contigs are slightly longer than the scaffolds, but approximately equal.

## Genome annotation

Our annotations are still being proccessed by RAST/BASys.

Based off of just the information that our sequences tend to match Pseudomonas, we can still answer some of the questions. 

- Some relatives of Pseudomonas tend to do well in moist conditions, such as those which infect burn wounds and swimming pools.

- Could possibly be a human pathogen, as one of the most well-known/well-researched members of the genus is Pseudomonas aeruginosa, which is definitely a pathogen; if we had the annotations for our sequence, we might be able to examine some of the Sec pathways and try to see whether or not it's infectious.

(_"Protein secretion systems in Pseudomonas aeruginosa: A wealth of pathogenic weapons". _ Bleves, S., Viarre, V., Salacha, R., Michel, G.P., Filloux, A., Voulhoux, R., 2010. Int. J. Med. Microbiol. 300, 534–543.)

- This genome may or may not encode certain antibiotic resistances; we would need to check the annotations for certain resistance casettes (such as kanR, ampR, etc). Some especially dangerous (to humans; ie those which can't be treated very easily) strains are resistant to common antibiotics. These strains have proteins which (similar to some as discussed in lecture) have a higher frequency of mutations, allowing them to more easily obtain resistance.

(Wong A, Rodrigue N, Kassen R (September 2012). _"Genomics of adaptation during experimental evolution of the opportunistic pathogen Pseudomonas aeruginosa"_. PLoS Genetics. 8 (9): e1002928. doi:10.1371/journal.pgen.1002928. PMC 3441735. PMID 23028345.)
